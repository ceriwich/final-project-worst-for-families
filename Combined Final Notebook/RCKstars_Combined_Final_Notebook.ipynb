## **Most Common Fire Type**

```python
import pandas as pd
import numpy as np
import math 
%matplotlib inline
import matplotlib.pyplot as plt

#/////////////////////////////////////////////////////////

raw_dataset = pd.read_csv("https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02")
fire_incidents = pd.DataFrame(raw_dataset)
```

Above, I am importing the necessary libraries, reading the dataset, and initializing a variable with the main DataFrame.

Below, I am filtering just the addresses in the city of Pittsburgh, which allows me to used the 'neighborhood' column of the data set more effectively. I then created a data set that displays the frequency of each type of fire alarm. 

```python
idx = 0
for town in fire_incidents['address']:
    if not('PITTSBURGH' in town):
        fire_incidents = fire_incidents.drop(idx)
    idx = idx + 1
fire_incidents.index = range(len(fire_incidents))

fire_types = pd.DataFrame({"quantity" : pd.Series(fire_incidents['type_description'].value_counts())})
fire_types.plot(kind="bar")
```

Cooking fires are the most frequent alarm, which is already a bad sign for new families, who rely on a stable and safe kitchen for food prep.

Below, I am selecting only the neighborhoods that had a cook fire alarm. 

```python
type_procesing = pd.DataFrame({"cook_fires" : pd.Series(fire_incidents['type_description']), 'neighborhood': pd.Series(fire_incidents['neighborhood'])})
idx = 0
for incident in type_procesing['cook_fires']:
    if not('Cooking fire, confined to container' in incident):
        type_procesing = type_procesing.drop(idx)
    idx = idx + 1
type_procesing.index = range(len(type_procesing))
```

Then, I will find the top five neighborhoods with the largest amounts of cook fire alarms

```python
cook_fires_dict = {}
for town in type_procesing['neighborhood']:
    if not pd.isna(town):
        if town not in cook_fires_dict:
            cook_fires_dict[town] = 1
        else:
            cook_fires_dict[town] += 1
cook_fire_freq = pd.Series(cook_fires_dict).to_frame('total_number_of_cook_fires')
cook_fire_freq = cook_fire_freq.sort_values(axis = 0, by = 'total_number_of_cook_fires', ascending = False)
cook_fire_freq.head(5).plot(kind = 'bar')
```

## **Number of Alarms Per Neighborhood**

Next, I will look at the total amount of alarms per neighborhood. Below, I am initializing a dataset with just the alarms and neighborhood columns, before grouping the data by neighborhood

```python
alarm_dict = {'alarms' : pd.Series(fire_incidents['alarms']), 'neighborhood': pd.Series(fire_incidents['neighborhood'])}
alarms_per_neighborhood = pd.DataFrame(alarm_dict)
alarm_counts = alarms_per_neighborhood.groupby('neighborhood', axis=0).count()
```

The next cell takes the alarm counts and put them in order by frequency. 

```python
alarm_arr = []
for num in alarm_counts['alarms']:
    alarm_arr.append(num)
alarm_arr.sort()
```

Then, I reassemble the neighborhoods and their corresponding alarm counts by utilizing a dicitonary. I finally use that dictionary to construct a final data frame with the neighborhoods with the most reported alarms

```python
top_five = {}
top_five_names = []
alarm_idx = len(alarm_arr) - 1
limit = alarm_idx - 6

while alarm_idx > limit:
    name = alarm_counts[alarm_counts['alarms'] == alarm_arr[alarm_idx]].index[0]
    top_five_names.append(name)
    top_five[name] = alarm_arr[alarm_idx]
    alarm_idx = alarm_idx - 1
```

```python
top_five_alarms = pd.Series(top_five).to_frame('total_number_of_alarms')
print(top_five_alarms)
```

## **Breakdown of Fire Types per Top Five Neighborhoods**

Finally, I will take the neighborhoods with the most alarms, and find out which types of alarms are most frequent with each neighborhood. I began by creating a data frame with all the necessary columns.

```python
types_dict = {'alarms' : pd.Series(fire_incidents['alarms']), 
              'neighborhood': pd.Series(fire_incidents['neighborhood']),
             'types_of_alarms' : pd.Series(fire_incidents['type_description'])}
types = pd.DataFrame(types_dict)
```

The functions below filter the types of alarms for each neighborhood and eliminate any instances of 0 alarms

```python
def types_per_town(town_name,types):
    town_dict = {}
    idx = 0
    for town in types['neighborhood']:
        if town == town_name:
            if types['types_of_alarms'][idx] not in town_dict:
                town_dict[types['types_of_alarms'][idx]] = types['alarms'][idx]
            else:
                town_dict[types['types_of_alarms'][idx]] += types['alarms'][idx]
        idx = idx + 1
    return town_dict

def del_zero_alarms(town_dict):
    to_del = []
    for k,v in town_dict.items():
        if v == 0:
            to_del.append(k)
    for n in to_del:
        town_dict.pop(n)
    return town_dict
```

Finally, we have they type and ferquency of alarms for the top eight neighbothoods with the most fire alarm reports. 

```python
for n in top_five_names:
    town_dict = del_zero_alarms(types_per_town(n, types))
    town = pd.Series(town_dict).to_frame('quanity')
    town.plot(kind = 'bar', title = f"Type and Frequency of Fire Alarms in {n} ")
    #print(f"Total number of alarms in {n} is {top_five_alarms['total_number_of_alarms'][n]}")
```

In this section I looked at the fire incident data for the neighborhoods of the city of Pittsburgh. First, I found the most common fire incident, which was cooking fires. The significance of this finding is in its relation to a new family's reliance on a safe kitchen. Firearms are seized by the Pittsburgh Bureau of Police for reasons such as suspicion of being used in a criminal offense, officer safety, and public safety. Guns not being seized from an area could potentially represent a lack of safety but for the purposes of this project, the neighborhoods with the most firearms seized are labeled as more dangerous. My thought process is that if there are more firearms being outwardly used in a way that they would be seized in the first place, this is more dangerous than if someone is hiding a bunch of firearms and not necesarrily doing anything with them. 


```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas
```

## This dataframe includes data from 2018 to 2022. The dataset was organized by address, not neighborhood, so some of the neighborhoods repeat.

```python
firearms_seized = pd.read_csv("firearms-seized.csv", index_col=None)
firearms_seized = firearms_seized.iloc[1792:, [1,7,10]]
firearms_seized = firearms_seized.dropna()
firearms_seized
```

## This list contains all of the neighborhoods in the dataset, sorted alphabetically. Duplicates are not included.

```python
neighborhood_list = firearms_seized["neighborhood"].tolist()
neighborhood_list.sort()

no_dups = []
for i in neighborhood_list:
    if i not in no_dups:
        no_dups.append(i)
```

## This dictionary contains two keys, 'Neighborhood' and 'Firearms Seized From 2018 to 2022'. The values for each of these keys are lists containing the respective data. At this point, the number of firearms seized from each neighborhood is correct, because the data from the same neighborhoods has been combined.

```python
neighborhood = []
total_firearms_seized = []
for item in no_dups:
    query_mask = firearms_seized["neighborhood"] == item
    this_neighborhood = firearms_seized[query_mask]
    total_sum = this_neighborhood["total_count"].sum()
    neighborhood.append(str(item))
    total_firearms_seized.append((total_sum))

dictionary = dict()
dictionary['Neighborhood'] = neighborhood
dictionary['Firearms Seized From 2018 to 2022'] = total_firearms_seized
```

## This dataframe was created from the dictionary above. It only lists each neighborhood once, with the correct total of firearms seized from it.

```python
new_dataframe = pd.DataFrame(dictionary, columns =['Neighborhood', 'Firearms Seized From 2018 to 2022'])
new_dataframe = new_dataframe.sort_values('Firearms Seized From 2018 to 2022')
new_dataframe
```

## This bar chart shows the whole data set from above. Because this is somewhat difficult to look at, I made another bar chart containing six neighborhoods with the highest number of firearms seized.

```python
new_dataframe.plot.bar(x = 'Neighborhood',
                       y = 'Firearms Seized From 2018 to 2022',
                       title = 'Firearms Seized From Pittsburgh Neighborhoods 2018 to 2022',
                       figsize = (20,10),
                       legend = None,
                       color = '#cea4cf')
```

## This bar chart shows the six neighborhoods in Pittsburgh with the highest number of firearms seized from 2018 to 2022.

```python
five_largest = new_dataframe.nlargest(5, 'Firearms Seized From 2018 to 2022')
five_largest = five_largest.sort_values('Firearms Seized From 2018 to 2022')
five_largest.plot.bar(x='Neighborhood',
                      y='Firearms Seized From 2018 to 2022',
                      title='Firearms Seized From Pittsburgh Neighborhoods 2018 to 2022',
                      figsize=(10, 8),
                      legend=None,
                      color='#cea4cf')
```

## This map shows all 87 neighborhoods from the dataframe. We can easily see that the neighborhood with the most firearms seized, Homewood South, has a lot more than the other neighborhoods.

```python
pittsburgh = geopandas.read_file("Neighborhoods/Neighborhoods_.shp") # read in the shapefile
pmap = pittsburgh.merge(new_dataframe, left_on='hood', right_on='Neighborhood')
pmap[['hood','Firearms Seized From 2018 to 2022','geometry']].head()
pmap = pmap.plot(column='Firearms Seized From 2018 to 2022',
          cmap='PuRd',
          edgecolor='white',
          figsize=(15, 10),
          legend=True,
          legend_kwds={'label': 'Number of Firearms', 'orientation' : 'vertical'})
pmap.set_axis_off()
pmap.set_title('Firearms Seized From Pittsburgh Neighborhoods 2018 to 2022')
pmap.plot()
```

## **The Best Neighborhood:**
The data above shows that the top five best neighborhoods for families to stay away from are:
1. Homewood South (171)
2. South Side Flats (99)
3. East Liberty (96)
4. Carrick (93)
5. Brighton Heights (93)

These neighborhoods have the most instances of criminal-related activity involving guns in which they are able to be siezed by the police. Therefore, it is safe to say that these may not be the most friendly areas for new families with young children.

## **Conclusion:** 
Based on the dataset, "Pittsburgh Police Firearm Seizures", the ultimate best neighborhood for new families to stay away from is Homewood South. This neighborhood was the only one to break 100 firearms seized from 2018 to 2022.
